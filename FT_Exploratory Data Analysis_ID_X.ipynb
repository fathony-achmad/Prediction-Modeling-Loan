## Term by Loan Status

```python
term_counts = df['term'].value_counts()
term_counts
```

```python
ax = sns.countplot(data=df, y='term', hue='loan_status')
for cont in ax.containers:
  ax.bar_label(cont)
```

```python
term_counts = df.groupby(['term', 'loan_status'])['int_rate'].count().reset_index()
```

```python
def calculate_ratio(group):
  total_int_rate = group['int_rate'].sum()
  group['ratio'] = round(group['int_rate'] / total_int_rate * 100, 0).astype(int)
  return group

term_counts = term_counts.groupby('term').apply(calculate_ratio)
```

```python
term_counts
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

ax[0].set_title('Status Loan by Term\n', fontsize=14)
sns.countplot(x='term', data=df, hue='loan_status', palette=['#ADBC9F', '#FF6868'], ax=ax[0])
for i, container in enumerate(ax[0].containers):
    for bar in container:
        bar_height = bar.get_height()
        text_pos = bar_height - 30
        ax[0].annotate(f'{int(bar_height):,}', (bar.get_x() + bar.get_width() / 2, text_pos), ha='center', va='center', fontsize=14)
ax[0].set_yticks([])
ax[0].set_xlabel('')
ax[0].set_ylabel('')
ax[0].legend([])

ax[1].set_title('Ratio of Status Loan by Term\n', fontsize=14)
term_counts.pivot(index='term', columns='loan_status', values='ratio').plot(kind='barh', stacked=True, color=['#ADBC9F', '#FF6868'], ax=ax[1])
for p in ax[1].patches:
    left, bottom, width, height = p.get_bbox().bounds
    if width != 0:
        ax[1].annotate((str(round(width, 2))+'%'), xy=(left+width/2, bottom+height/2), ha='center', va='center', fontsize=14)
ax[1].set_xticks([])
ax[1].set_xlabel('')
ax[1].set_ylabel('')
ax[1].legend(loc='upper left', bbox_to_anchor=(1.02, 1), fontsize='12')

plt.tight_layout()
plt.show()
```



*   Jangka waktu pinjaman ('term') memberi tahu kami tentang jumlah pembayaran pinjaman.
*   jumlah pinjaman untuk jangka waktu 36 bulan (151.101) lebih besar daripada jangka waktu 60 bulan.
*   Pinjaman 60 bulan, rasio "good" adalah 36.0% dan rasio "bad" adalah 64.0%.
Sedangkan untuk pinjaman 36 bulan, rasio "good" adalah 18.0% dan rasio "bad" adalah 82.0%.
*   Pinjaman dengan jangka waktu 60 bulan memiliki kinerja yang lebih baik dibandingkan dengan pinjaman 36 bulan.

## Defalut loan_status

```python
# membuat fungsi rata-rata loan status default
def get_default_rate_by_col(column):
    """
    Param:
        column: str, column for which default rate is to be calculated for it's unique values. Column must have limited
        unique values.
    """
    tmd=pd.DataFrame(df[df['loan_status']=='Default'][column].value_counts()/df[column].value_counts())
    tmd.columns=[column+'_Default_Rate']
    tmd=tmd.join(df[column].value_counts())
    tmd.columns=[column+'_Default_Rate',column+'_Count']
    return tmd
```

```python
# Membuat issue_year yang berisi year in which the loan was issued.
df['issue_year']=df['issue_d'].apply(lambda x:int(x.split('-')[1]))
```

## Fully Paid loan_status

```python
# membuat fungsi rata-rata loan status fully paid
def get_fullypaid_rate_by_col(column):
    """
    Param:
        column: str, column for which Fully Paid rate is to be calculated for it's unique values. Column must have limited
        unique values.
    """
    tmfp=pd.DataFrame(df[df['loan_status']=='Fully Paid'][column].value_counts()/df[column].value_counts())
    tmfp.columns=[column+'_Fully Paid_Rate']
    tmfp=tmfp.join(df[column].value_counts())
    tmfp.columns=[column+'_Fully Paid_Rate',column+'_Count']
    return tmfp
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(dfc1)}')

filtered_entries = np.array([True] * len(dfc1))
for col in dfc1:
    if dfc1[col].dtype.kind in ['i', 'f']:  # Check for numeric data types
        Q1 = dfc1[col].quantile(0.25)
        Q3 = dfc1[col].quantile(0.75)
        IQR = Q3 - Q1
        low_limit = Q1 - (IQR * 1.5)
        high_limit = Q3 + (IQR * 1.5)
        filtered_entries = ((dfc1[col] >= low_limit) & (dfc1[col] <= high_limit)) & filtered_entries
    else:
        print(f"Skipping column {col} (non-numeric)")  # Optional: Inform about skipped columns

dfc1_filter = dfc1[filtered_entries]

print(f'Jumlah baris setelah memfilter outlier: {len(dfc1_filter)}')
```

```python
def top_correlation(dfc1_fe, n):
    corr_matrix = dfc1_fe.corr()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    correlation = (upper.stack()
                   .sort_values(ascending=False))
    correlation = pd.DataFrame(correlation).reset_index()
    correlation.columns = ["Variable_1", "Variable_2", "Correlation"]
    correlation = correlation.reindex(correlation['Correlation'].abs().sort_values(ascending=False).index).reset_index(drop=True)
    return correlation.head(n)

print("High Correlated Features (Corr > 0.5)")
top_correlation(uni_dist, 39)
```
